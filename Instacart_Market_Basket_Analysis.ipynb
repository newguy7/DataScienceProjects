```python
import pandas as pd

# Load datasets
aisles_df = pd.read_csv('aisles.csv')
departments_df = pd.read_csv('departments.csv')
order_products_prior_df = pd.read_csv('order_products__prior.csv')
order_products_train_df = pd.read_csv('order_products__train.csv')
orders_df = pd.read_csv('orders.csv')
products_df = pd.read_csv('products.csv')
```

```python
# Display the first few rows to understand structure
print("table1: aisles")
print(aisles_df.head())
print("\n")
print("table2: departments")
print(departments_df.head())
print("\n")
print("table3: order_prior")
print(order_products_prior_df.head())
print("\n")
print("table4: order_train")
print(order_products_train_df.head())
print("\n")
print("table5: orders")
print(orders_df.head())
print("\n")
print("table6: products")
print(products_df.head())
```

```python
# Merging orders with prior order products
order_products_prior_merged = pd.merge(order_products_prior_df, orders_df, on='order_id')
order_products_prior_merged = pd.merge(order_products_prior_merged, products_df, on='product_id')
```

```python
order_products_prior_merged.head()
```

```python
print(order_products_prior_merged.shape)
```

```python
#Calculate product reorder frequency

product_reorder_freq = order_products_prior_merged.groupby('product_id').size().reset_index(name='reorder_count')
product_reorder_freq.head()
```

```python
# Calculate user-specific product preferences
user_product_pref = order_products_prior_merged.groupby(['user_id', 'product_id','reordered']).size().reset_index(name='user_product_count')
user_product_pref[user_product_pref['user_id']==1].sort_values(by='user_product_count',ascending=False)
```
